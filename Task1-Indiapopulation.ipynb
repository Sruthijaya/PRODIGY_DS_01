OBJECTIVE

This analysis focuses on the demographic changes in India from 1955 to 2020
OBJECTIVES
- To explore and visualize the urban population percentage in India over the years.
- To analyze the distribution of the median age in India over the same period.
- To interpret the correlation between various demographic variables to understand their relationships and impacts.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('india_population.csv')
data.head()
```

```python
data.corr()
```

```python
# Plotting Correlation Matrix
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

- Population and Year have Strong positive correlation (1), indicating that population has increased steadily over the years.
- Year and Median Age have High positive correlation (0.92), suggesting that the median age has risen as time progresses.
- Population and Urban Population have tremely high positive correlation (0.99), showing that as the total population grows, the urban population also increases.
- Fertility Rate and Year have Strong negative correlation (-0.99), indicating that fertility rates have declined over the years.
- Median Age and Fertility Rate have Strong negative correlation (-0.95), suggesting that as the population ages, fertility rates drop.
- Yearly % Change and Population have Negative correlation (-0.82), showing that yearly population growth rates tend to decrease as the total population increases.

```python
# Bar Chart-Urban Population Percentage Over the Years
plt.figure(figsize=(10, 6))
plt.bar(data['Year'], data['Urban Pop %'], color='thistle')
plt.xlabel('Year')
plt.ylabel('Urban Population Percentage')
plt.title('Urban Population Percentage in India Over the Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The urban population percentage has consistently risen over the decades, indicating significant urbanization.
There is a notable steep increase in urbanization after the year 2000, reflecting accelerated urban development.
By 2020, more than 35% of India's population resided in urban areas.

```python
# Histogram - Distribution of Median Age
plt.figure(figsize=(10, 6))
plt.hist(data['Median Age'], bins=10, color='lightcoral', edgecolor='black')
plt.xlabel('Median Age')
plt.ylabel('Frequency')
plt.title('Distribution of Median Age in India')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The median age of 20 appears most frequently, indicating that for several years, the median age in India was around 20.There is also a significant frequency at the median age of 28, reflecting a trend towards an aging population in recent years.

CONCLUSION

There has been a marked increase in the urban population percentage, reflecting rapid urbanization and economic development.
The median age has increased, indicating an aging population due to factors such as reduced fertility rates and increased life expectancy.
Strong correlations exist between various demographic variables, highlighting the interconnected nature of population dynamics.
